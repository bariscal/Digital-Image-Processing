```python
from PIL import Image
import numpy as np
import matplotlib.image as mpimg
import matplotlib.pyplot as plt

import cv2
```

In first line, I read the image file with openCV. In the second line, I converted image datas type to 32-bit. That code does not change the values of array. In third line, I changed values. In 8-bit RGB, values are between [0,255] but in 32-bit coding, RGB values are changing in between [0,4294967295]. I found this value with this operation:

I multiply 8-bit values with 2^24 (256x256x256) and I created new array with 32-bit unsigned integer. Arrays and data types can be seen in results.

```python
i = cv2.imread('ginger8Bit.jpg') # Need to be sure to have a 8-bit input
img = np.array(i, dtype=np.uint32) # This line only change the type, not values
img *= 256*256*256
# Now we get the good values in 16 bit format
```
